1.Read an image and convert it into gray scale image without using built-in function.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Read the image (color image)
img = cv2.imread("/content/cat.jpg")   # Put your image name here

# Step 2: Get image dimensions
height, width, channels = img.shape

# Step 3: Create an empty grayscale image
gray_img = np.zeros((height, width), dtype=np.uint8)

# Step 4: Convert RGB to Grayscale manually
for i in range(height):
    for j in range(width):
        B = img[i, j, 0]
        G = img[i, j, 1]
        R = img[i, j, 2]

        gray_value = int(0.299*R + 0.587*G + 0.114*B)
        gray_img[i, j] = gray_value

# Step 5: Display images
plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
plt.title("Original Image")
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.axis("off")

plt.subplot(1,2,2)
plt.title("Grayscale Image (Manual)")
plt.imshow(gray_img, cmap='gray')
plt.axis("off")

plt.show()
```
