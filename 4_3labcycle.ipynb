4.Display the histogram of the gray scale image.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Read the image
img = cv2.imread("/content/dog.jpg")

# Step 2: Convert to grayscale manually
height, width, _ = img.shape
gray = np.zeros((height, width), dtype=np.uint8)

for i in range(height):
    for j in range(width):
        B, G, R = img[i, j]
        gray[i, j] = int(0.299*R + 0.587*G + 0.114*B)

# Step 3: Initialize histogram array (256 intensity levels)
hist = np.zeros(256, dtype=int)

# Step 4: Calculate histogram manually
for i in range(height):
    for j in range(width):
        intensity = gray[i, j]
        hist[intensity] += 1

# Step 5: Display grayscale image and histogram
plt.figure(figsize=(12,4))

plt.subplot(1,2,1)
plt.title("Grayscale Image")
plt.imshow(gray, cmap='gray')
plt.axis("off")

plt.subplot(1,2,2)
plt.title("Histogram of Grayscale Image")
plt.plot(hist)
plt.xlabel("Intensity Value")
plt.ylabel("Number of Pixels")

plt.show()
```
