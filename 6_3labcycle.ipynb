6.Display the edge map of an image with any edge detection algorithm.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Read the image
img = cv2.imread("/content/cat.jpg")

# Step 2: Convert to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Step 3: Apply Sobel operator
sobel_x = cv2.Sobel(gray, cv2.CV_64F, 1, 0, ksize=3)
sobel_y = cv2.Sobel(gray, cv2.CV_64F, 0, 1, ksize=3)

# Step 4: Combine X and Y edges
edge_map = np.sqrt(sobel_x**2 + sobel_y**2)
edge_map = np.uint8(edge_map)

# Step 5: Display results
plt.figure(figsize=(12,4))

plt.subplot(1,3,1)
plt.title("Original Image")
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.axis("off")

plt.subplot(1,3,2)
plt.title("Grayscale Image")
plt.imshow(gray, cmap='gray')
plt.axis("off")

plt.subplot(1,3,3)
plt.title("Edge Map (Sobel)")
plt.imshow(edge_map, cmap='gray')
plt.axis("off")

plt.show()
```
