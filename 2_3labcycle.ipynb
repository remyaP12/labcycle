Read an image and display the RGB channels of the images separately

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Read the image
img = cv2.imread("/content/2dog.jpg")   # Replace with your image name

# Step 2: Convert BGR to RGB (OpenCV reads image in BGR format)
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Step 3: Separate channels
R = img_rgb[:, :, 0]
G = img_rgb[:, :, 1]
B = img_rgb[:, :, 2]

# Step 4: Display original image and RGB channels
plt.figure(figsize=(12,6))

plt.subplot(1,4,1)
plt.title("Original Image")
plt.imshow(img_rgb)
plt.axis("off")

plt.subplot(1,4,2)
plt.title("Red Channel")
plt.imshow(R, cmap='gray')
plt.axis("off")

plt.subplot(1,4,3)
plt.title("Green Channel")
plt.imshow(G, cmap='gray')
plt.axis("off")

plt.subplot(1,4,4)
plt.title("Blue Channel")
plt.imshow(B, cmap='gray')
plt.axis("off")

plt.show()
```
